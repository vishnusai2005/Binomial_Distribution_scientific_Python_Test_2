**1. Find the probability distribution (Probability of each outcome) for getting a five when rolling a dice four times.**

```
- Output Format:
Probability distribution for getting a five when rolling a dice 4 times:
P(X=0): ______
P(X=1): ______
P(X=2): ______
P(X=3): ______
P(X=4): ______
```

````python
print("Probability distribution for getting a five when rolling a dice 4 times:")
for i in range(0,5):
  print(f"P(X={i}): {binom.pmf(i,4,1/6):.4f}")
````

**2. A fair coin is tossed 10 times. Find out the probability of:**
1. Getting a head at most 6 times
2. Getting a head at least 3 times

```
- Output Format:
Coin Toss Probabilities:
Probability of getting at most 6 heads: ______
Probability of getting at least 3 heads: ______
```

````python
print("Coin Toss Probabilities:")
print(f"Probability of getting at most 6 heads: {binom.cdf(6,10,0.5):.4f}")
print(f"Probability of getting at least 3 heads: {1 - binom.cdf(2,10,0.5):.4f}")
````

**3. A surgeon claims that the success rate of surgeries performed by him is 65%. What is the probability that out of next 100 patients:**
1. At least 60 patients will have a successful surgery
2. Fewer than 75 patients will have a successful surgery
3. There will be no more than 30 failed surgeries

```
- Output Format:
Surgery Success Rate Probabilities:
Probability of at least 60 successful surgeries: _____
Probability of less than 75 successful surgeries: _____
Probability of no more than 30 failed surgeries: _____
```

````python
print("Surgery Success Rate Probabilities:")
print(f"Probability of at least 60 successful surgeries: {1-binom.cdf(59,100,0.65):.4f}")
print(f"Probability of less than 75 successful surgeries: {binom.cdf(74,100,0.65):.4f}")
print(f"Probability of no more than 30 failed surgeries: {binom.cdf(30,100,0.35):.4f}")
````

**5. A factory manufactures car batteries with a defect rate of 1.5%. In each shipment to a car company there are 10000 batteries. The cost paid for each battery by the the car company is 50000 rupees and the battery manufacturer does not provide any compensation for the defective batteries found in a shipment. Calculate the proability that the loss (in rupees) by the car company for a single shipment:**
1. Will not be more than 7500000
2. Will be between 6000000 to 9000000
3. Will be between 7000000 to 8000000

```
- Output Format:
Car Battery Defect Loss Probabilities:
Probability that loss will not be more than 7,500,000: ____
Probability that loss will be between 6,000,000 and 9,000,000: ____
Probability that loss will be between 7,000,000 and 8,000,000: ____
```

````python
print("Car Battery Defect Loss Probabilities:")
print(f"Probability that loss will not be more than 7,500,000: {binom.cdf(7500000//50000,10000,0.015):.4f}")
print(f"Probability that loss will be between 6,000,000 and 9,000,000: {binom.cdf(9000000//50000,10000,0.015) - binom.cdf((6000000//50000)-1,10000,0.015):.4f}")
print(f"Probability that loss will be between 7,000,000 and 8,000,000: {binom.cdf(8000000//50000,10000,0.015) - binom.cdf((7000000//50000) - 1, 10000, 0.015):.4f}")
````
